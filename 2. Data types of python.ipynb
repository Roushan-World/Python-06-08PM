# 2. Data types of Python

```python
#1. String
print('Hello world')
print('Jayshree')

#2. Integer
print(12)


#3. Float 
print(12.5)
print(100.0)

#4. Boolean

print(True)

print(False)
print(bool(1))   # output = True
print(bool(0))   # output= False
```

# type() function 

```python
# 1. String

print(type('Hello world'))
print(type('Jayshree'))

#2. Integer

print(type(12))
print(type(100))

#3. Float

print(type(100.0))
print(type(112.2333333))

#4. Boolean

print(type(True))
print(type(False))
print(type(bool(1)))
print(type(bool(0)))
```

```python
# lets convert the integer number 10 into a string and float.

number=100

print(str(number))
print(type(str(number)))
print(float(number))

print(type(float(number)))
```

```python
# Let's convert the float number 3.14 in to a string and integer.

number=3.14

print(str(number))
print(type(str(number)))

print(int(number))
print(type(int(number)))
```

```python
# let's convert the boolean to an integer, a float and a string.

number_1=True
number_2=False

print(int(number_1))
print(type(int(number_1)))

print(int(number_2))
print(type(int(number_2)))

print(float(number_1))
print(type(float(number_1)))

print(float(number_2))
print(type(float(number_2)))

print(str(number_1))
print(type(str(number_1)))
print(str(number_2))
print(type(str(number_2)))


print(number_1)
print(type(number_1))
print(number_2)
print(type(number_2))
```

```python
# Let's find the data type of 9/3 and 9//4
print(9/3)  # Float divison
print(9//3) # Integer divison

print(9/4)  # Float Divison
print(9//4) # Integer Divison
```

```python
# Multiplication

x=8*7
print(x)
print(type(x))
```

```python
# Divison

x = 125/24
print(x)
print(type(x))
```

```python
# Floor divison

x = 125//24
print(x)
print(type(x))
```

```python
# Modulus

x = 125%24
print(x)
print(type(x))
```

```python
# Exponent

x = 2**3
print(x)
print(type(x))
```

```python
# An Example: How many minutes there are in 20 hrs?

one_hrs=60  # 60 minutes

hour= 20

minutes=one_hrs*hour

print(minutes)
print(type(minutes))
```

```python
# An Example: Let's calculate How many hours are there are in 348 minutes.

minutes= 348 # 60 Second

one_hrs=60

hours=minutes/one_hrs

print(hours)
print(type(hours))

```
